```python
import matplotlib.pyplot as plt
```

```python
def show_result(labels, score, title, color):
    tick = range(len(labels))
    bar = plt.bar(tick, score)
    for i in tick:
        bar[i].set_color(color)
    plt.xticks(tick, labels)
    plt.title(title)
    plt.xlabel('algorithm')
    plt.ylabel('accuracy')
    plt.show()
```

```python
# 6个类
labels = ['bagging', 'smote', 'under-sampling', 'under+over']
score = [0.707, 0.717, 0.62, 0.717]

show_result(labels, score, '6 class')
```

```python
# 5个类
labels = ['bagging', 'naive bayes', 'SMOTE']
score = [0.811, 0.779, 0.766]

show_result(labels, score, '5 class')
```

```python
# 4个类
labels = ['bagging', 'naive bayes', 'SMOTE']
score = [0.866, 0.828, 0.812]

show_result(labels, score, '4 class', 'g')
```

```python
# 全部放在一起对比

labels = ['bagging', 'naive bayes', 'SMOTE']
# 6个类
score = [0.707, 0.687, 0.717]

# 5个类
score5 = [0.811, 0.779, 0.766]

# 4个类
score4 = [0.866, 0.828, 0.812]

tick = range(len(labels))
tick5 = [i - 0.21 for i in tick]
tick4 = [i + 0.21 for i in tick]
plt.bar(tick, score, width=0.2)
plt.bar(tick5, score5, width=0.2)
plt.bar(tick4, score4, width=0.2)
plt.xticks(tick, labels)
plt.legend(('6 class', '5 class', '4 class'))
plt.title('compare result')
plt.xlabel('algorithm')
plt.ylabel('accuracy')
plt.ylim(0, 1.1)
plt.show()
```
