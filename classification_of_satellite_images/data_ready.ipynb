## 准备数据
将类别变成数字，去掉不需要的

```python
import pandas as pd
```

### 训练数据

```python
train_data = pd.read_csv('training.csv')
```

```python
# 指出类别到数字的转换
labels = set(train_data['class'])
i = 0
label_to_digit = {}
for label in labels:
    label_to_digit[label] = i
    i += 1
print(label_to_digit)
```

```python
# 替换
for k,v in label_to_digit.items():
    train_data = train_data.replace(k,v)
train_data.head()
```

```python
# 去掉第一行和第二列并写入文件
d = train_data.drop(['max_ndvi'],axis=1)
d.head()
d.to_csv('train_data.csv',index='class')
```
