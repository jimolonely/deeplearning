```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train_data.csv').values
test = pd.read_csv('test_data.csv').values

print(type(train))
# X_train,y_train = train[:,1:],train[:,0]
# # print(X_train,y_train)
# X_test,y_test = test[:,1:],test[:,0]
# # print(len(X_train[0])==len(X_test[0]))

# 取出类别
class_test = test[:, 0]
print(class_test.shape)

class_train = train[:, 0]
print(class_train.shape)
```

```python
def show_data(test, train):
    plt.figure(1, figsize=(20, 3))
    plt.scatter(train, class_train, c='y')
    plt.scatter(test, class_test)
    plt.show()
```

```python
show_data(test[:, 1], train[:, 1])
```

```python
show_data(test[:, 28], train[:, 28])
```
