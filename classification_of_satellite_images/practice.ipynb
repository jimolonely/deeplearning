```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('training.csv')
print(data.shape)
```

```python
class_set = set(data['class'])
print(class_set)
```

```python
farm = data[data['class']=='farm']
print(farm.shape)
```

```python
farm_labels = farm.T.index[2:].values
print(farm_labels)
```

```python
farm_data = farm.values[:,2:]
print(farm_data)
```

```python
x = range(len(farm_labels))
plt.title('farm')
for y in farm_data[:10,:]:
    plt.plot(x,y)
    plt.xticks(x,farm_labels,rotation=90)
plt.show()
```

## 将上述过程写成函数

```python
def draw_data(data,size):
    class_set = set(data['class'])
    row = len(class_set)/2
    col = 2
    i = 0
    plt.figure(1,figsize=(15,25))
    for c in class_set:
        cls = data[data['class']==c]
        label = cls.T.index[2:].values
        d = cls.values[:,2:]
        i+=1
        plt.subplot(row,col,i)
        x = range(len(label))
        plt.title(c)
        for y in d[:size,:]:
            plt.plot(x,y)
            plt.xticks(x,label,rotation=45)
    plt.show()
```

```python
draw_data(data,10)
```

## 直观查看二维数据

```python
def draw_data2(data,size):
    class_set = set(data['class'])
    row = len(class_set)/2
    col = 2
    i = -1
    color = ['r','g','b','m','y','k']
    plt.figure(1,figsize=(15,5))
    for c in class_set:
        cls = data[data['class']==c]
        label = cls.T.index[2:].values
        x = range(len(label))
        d = cls.values[:,2:]
        i+=1
        for y in d[:size,:]:
            plt.scatter(x,y,c=color[i])
    #plt.xticks(x,label,rotation=45)
    plt.show()
```

```python
draw_data2(data,100)
```
